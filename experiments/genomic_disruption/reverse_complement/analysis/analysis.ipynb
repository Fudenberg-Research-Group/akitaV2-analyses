```python
import warnings
warnings.filterwarnings("ignore") 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("/home1/smaruj/akitaX1-analyses/figures/plot_styles/global_plotting_style.mplstyle")
from scipy import stats

from helper import read_reverse_complement_disruption
```

### Reading disruption and reverse complement disruption scores

```python
disruption_dir = "/project/fudenber_735/akitaX1_analyses_data/genomic_disruption/disruption_by_permutation/"
rc_disruption_dir = "/project/fudenber_735/akitaX1_analyses_data/genomic_disruption/reverse_compliment/"

df = read_reverse_complement_disruption(disruption_dir, rc_disruption_dir)
```

### Disruption score is not strand-sensitive

```python
plt.figure(figsize=(8, 7))

plt.scatter(x=df["SCD"], y=df["RC_SCD"], alpha=0.5, edgecolors='w', linewidth=0.2, color='blue')
rc_corr = round(stats.pearsonr(df["SCD"], df["RC_SCD"]).statistic, 4)

plt.xlabel("Disruption Score, Default Strand")
plt.ylabel("Disruption Score, Reverse Complement")

plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)

plt.text(x=0.71, y=0.05, s=f"Pearson R = {rc_corr}", transform=plt.gca().transAxes, verticalalignment='top')

# plt.savefig("./plots/disruption_score_reverse_complement.pdf", format="pdf", bbox_inches="tight")
```

### Plots for insulation scores (INS-16, INS-64)

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x=df["INS-16"], y=df["RC_INS-16"])
rc_corr = round(stats.pearsonr(df["INS-16"], df["RC_INS-16"]).statistic, 4)
plt.title(f"Average INS-16 correlation between predictions for default strand \n and its reverse complement, corr={rc_corr}")
plt.xlabel(f"ave INS-16, default strand")
plt.ylabel(f"ave INS-16, random complement")
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x=df["INS-64"], y=df["RC_INS-64"])
rc_corr = round(stats.pearsonr(df["INS-64"], df["RC_INS-64"]).statistic, 4)
plt.title(f"Average INS-64 correlation between predictions for default strand \n and its reverse complement, corr={rc_corr}")
plt.xlabel(f"ave INS-64, default strand")
plt.ylabel(f"ave INS-64, random complement")
```
