## Disruption scores for specific sites are not strand-sensitive

```python
plt.figure(figsize=(8, 7))

plt.scatter(x=df["SCD"], y=df["RC_SCD"], alpha=0.5, edgecolors='w', linewidth=0.2, color='blue')
rc_corr = round(stats.pearsonr(df["SCD"], df["RC_SCD"]).statistic, 4)

plt.xlabel("Disruption Score - Default Strand")
plt.ylabel("Disruption Score - Reverse Complement")

plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)

plt.text(x=0.71, y=0.05, s=f"Pearson R = {rc_corr}", transform=plt.gca().transAxes, verticalalignment='top')

plt.savefig("disruption_score_reverse_complement.pdf", format="pdf", bbox_inches="tight")
```

```python
plt.figure(figsize=(6,5))
sn.scatterplot(x=df["INS-16"], y=df["RC_INS-16"])
rc_corr = round(stats.pearsonr(df["INS-16"], df["RC_INS-16"]).statistic, 4)
plt.title(f"Average INS-16 correlation between predictions for default strand \n and its reverse complement, corr={rc_corr}")
plt.xlabel(f"ave INS-16, default strand")
plt.ylabel(f"ave INS-16, random complement")
```
