```python
df_R = read_multi_model_data(data_dir_R, keys_to_ignore = ["disruption_SCD", "insertion_SCD"])
```

```python
spacing_df_R = df_R.groupby(["spacer_bp"]).agg({"SCD": "mean"}).reset_index()
spacing_df_L = df_L.groupby(["spacer_bp"]).agg({"SCD": "mean"}).reset_index()
spacing_df_C = df_C.groupby(["spacer_bp"]).agg({"SCD": "mean"}).reset_index()
spacing_df_D = df_D.groupby(["spacer_bp"]).agg({"SCD": "mean"}).reset_index()
```

```python
# distinct colors for different orientations
convergent_color = "#1f77b4" #Vivid Cyan-Blue
divergent_color = "#ff7f0e" #Bright Orange
left_color = "#2ca02c" #Crisp Green
right_color = "#d62728" #Rich Magenta
```

## Raw plots for all orientations

```python
plt.figure(figsize=(10,8))
x = [2 * spacer + 60 for spacer in spacing_df_R.spacer_bp.unique()]

plt.plot(x, spacing_df_R["SCD"], c=right_color, label="right >>")
plt.plot(x, spacing_df_L["SCD"], c=left_color, label="left <<")
plt.plot(x, spacing_df_C["SCD"], c=convergent_color, label="convergnet ><")
plt.plot(x, spacing_df_D["SCD"], c=divergent_color, label="divergent <>")

plt.legend()
plt.xlabel("Inter-motif Distance")
plt.ylabel("Insertion Score")
```

## Smoothed plots for all orientations

```python
bin_size = 19
```

```python
def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_
```

```python
R_percentiles_df = df_R.groupby(["spacer_bp"])["SCD"].agg(
    percentile25 = percentile(25), percentile75 = percentile(75)).reset_index()
```

```python
L_percentiles_df = df_L.groupby(["spacer_bp"])["SCD"].agg(
    percentile25 = percentile(25), percentile75 = percentile(75)).reset_index()
```

```python
C_percentiles_df = df_C.groupby(["spacer_bp"])["SCD"].agg(
    percentile25 = percentile(25), percentile75 = percentile(75)).reset_index()
```

```python
D_percentiles_df = df_D.groupby(["spacer_bp"])["SCD"].agg(
    percentile25 = percentile(25), percentile75 = percentile(75)).reset_index()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,8))
x = [2*spacer + 60 for spacer in spacing_df_R.spacer_bp.unique()]

# right
plt.plot(x, numutils.smooth(spacing_df_R["SCD"], bin_size), c=right_color, label="right >>")
R_percentile25_line = numutils.smooth(R_percentiles_df["percentile25"], bin_size)
R_percentile75_line = numutils.smooth(R_percentiles_df["percentile75"], bin_size)
ax.fill_between(x, R_percentile25_line, R_percentile75_line, alpha=0.20, color=right_color) 

# left
plt.plot(x, numutils.smooth(spacing_df_L["SCD"], bin_size), c=left_color, label="left <<")
L_percentile25_line = numutils.smooth(L_percentiles_df["percentile25"], bin_size)
L_percentile75_line = numutils.smooth(L_percentiles_df["percentile75"], bin_size)
ax.fill_between(x, L_percentile25_line, L_percentile75_line, alpha=0.20, color=left_color) 

# convergent
plt.plot(x, numutils.smooth(spacing_df_C["SCD"], bin_size), c=convergent_color, label="convergent ><")
C_percentile25_line = numutils.smooth(C_percentiles_df["percentile25"], bin_size)
C_percentile75_line = numutils.smooth(C_percentiles_df["percentile75"], bin_size)
ax.fill_between(x, C_percentile25_line, C_percentile75_line, alpha=0.20, color=convergent_color) 

# divergent
plt.plot(x, numutils.smooth(spacing_df_D["SCD"], bin_size), c=divergent_color, label="divergent <>")
D_percentile25_line = numutils.smooth(D_percentiles_df["percentile25"], bin_size)
D_percentile75_line = numutils.smooth(D_percentiles_df["percentile75"], bin_size)
ax.fill_between(x, D_percentile25_line, D_percentile75_line, alpha=0.20, color=divergent_color) 

plt.legend(loc="upper right")
plt.xlabel("Inter-motif Distance")
plt.ylabel("Insertion Score")
# plt.savefig("./plots/insertion_score_vs_intermotif_distance.pdf", format="pdf", bbox_inches="tight")
```

# Zoom-in into 1-5kb

```python
zoom_distance = 5000
flank_length = 30
```

```python
inter_motif_dist = (zoom_distance-(flank_length*2))//2
```

```python
spacing_df_R_zoom = spacing_df_R[spacing_df_R["spacer_bp"] <= inter_motif_dist]
spacing_df_L_zoom = spacing_df_L[spacing_df_L["spacer_bp"] <= inter_motif_dist]
spacing_df_C_zoom = spacing_df_C[spacing_df_C["spacer_bp"] <= inter_motif_dist]
spacing_df_D_zoom = spacing_df_D[spacing_df_D["spacer_bp"] <= inter_motif_dist]
```

```python
print("Distance giving minimum in the first 5kb")
print("Orientation: RIGHT")
smoothed_R = numutils.smooth(spacing_df_R_zoom["SCD"], bin_size)
print("\t inter-motif distance = ", int(smoothed_R[smoothed_R.argmin() * 2]))
print("\t with flanks =", int(smoothed_R[smoothed_R.argmin() * 2]) * 2 + flank_length*2)

print("Orientation: LEFT")
smoothed_L = numutils.smooth(spacing_df_L_zoom["SCD"], bin_size)
print("\t inter-motif distance = ", int(smoothed_L[smoothed_L.argmin() * 2]))
print("\t with flanks =", int(smoothed_L[smoothed_L.argmin() * 2]) * 2 + flank_length*2)

print("Orientation: CONVERGENT")
smoothed_C = numutils.smooth(spacing_df_C_zoom["SCD"], bin_size)
print("\t inter-motif distance = ", int(smoothed_C[smoothed_C.argmin() * 2]))
print("\t with flanks =", int(smoothed_C[smoothed_C.argmin() * 2]) * 2 + flank_length*2)

print("Orientation: DIVERGENT")
smoothed_D = numutils.smooth(spacing_df_D_zoom["SCD"], bin_size)
print("\t inter-motif distance = ", int(smoothed_D[smoothed_D.argmin() * 2]))
print("\t with flanks =", int(smoothed_D[smoothed_D.argmin() * 2]) * 2 + flank_length*2)
```

```python
R_percentiles_df_zoom = R_percentiles_df[R_percentiles_df["spacer_bp"] <= inter_motif_dist]
L_percentiles_df_zoom = L_percentiles_df[L_percentiles_df["spacer_bp"] <= inter_motif_dist]
C_percentiles_df_zoom = C_percentiles_df[C_percentiles_df["spacer_bp"] <= inter_motif_dist]
D_percentiles_df_zoom = D_percentiles_df[D_percentiles_df["spacer_bp"] <= inter_motif_dist]
```
