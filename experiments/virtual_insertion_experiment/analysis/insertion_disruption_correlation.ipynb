### Disruption data

```python
dis_stat_names = ["SCD", "SSD", "INS-16", "INS-64"]
```

```python
dis_data_dir = "/project/fudenber_735/akitaX1_analyses_data/genomic_disruption_experiment/disruption_by_permutation"
```

```python
# reading h5 files to dataframes
dis_df_m0 = h5_to_df(dis_data_dir+"/model_0.h5", dis_stat_names, average=False) 
dis_df_m1 = h5_to_df(dis_data_dir+"/model_1.h5", dis_stat_names, average=False) 
dis_df_m2 = h5_to_df(dis_data_dir+"/model_2.h5", dis_stat_names, average=False) 
dis_df_m3 = h5_to_df(dis_data_dir+"/model_3.h5", dis_stat_names, average=False) 
dis_df_m4 = h5_to_df(dis_data_dir+"/model_4.h5", dis_stat_names, average=False) 
dis_df_m5 = h5_to_df(dis_data_dir+"/model_5.h5", dis_stat_names, average=False) 
dis_df_m6 = h5_to_df(dis_data_dir+"/model_6.h5", dis_stat_names, average=False) 
dis_df_m7 = h5_to_df(dis_data_dir+"/model_7.h5", dis_stat_names, average=False) 
```

```python
dis_df_m0 = average_stat_over_targets(dis_df_m0, model_index=0, head_index=1)
dis_df_m1 = average_stat_over_targets(dis_df_m1, model_index=1, head_index=1)
dis_df_m2 = average_stat_over_targets(dis_df_m2, model_index=2, head_index=1)
dis_df_m3 = average_stat_over_targets(dis_df_m3, model_index=3, head_index=1)
dis_df_m4 = average_stat_over_targets(dis_df_m4, model_index=4, head_index=1)
dis_df_m5 = average_stat_over_targets(dis_df_m5, model_index=5, head_index=1)
dis_df_m6 = average_stat_over_targets(dis_df_m6, model_index=6, head_index=1)
dis_df_m7 = average_stat_over_targets(dis_df_m7, model_index=7, head_index=1)
```

```python
disruptionSCD_collected = pd.concat([dis_df_m0["SCD_m0"], dis_df_m1["SCD_m1"], dis_df_m2["SCD_m2"], dis_df_m3["SCD_m3"], dis_df_m4["SCD_m4"], dis_df_m5["SCD_m5"], dis_df_m6["SCD_m6"], dis_df_m7["SCD_m7"]], axis=1)
```

```python
# averaging over models
model_numbers = 8
disruptionSCD_collected["SCD"] = disruptionSCD_collected[[f"SCD_m{model_index}" for model_index in range(model_numbers)]].mean(axis=1)
insertionSCD_collected["SCD"] = insertionSCD_collected[[f"SCD_m{model_index}" for model_index in range(model_numbers)]].mean(axis=1)
```

## Correlation between insertion and disruption SCD

```python
sns.scatterplot(x=disruptionSCD_collected["SCD"], y=insertionSCD_collected["SCD"])
perm_corr = round(stats.pearsonr(disruptionSCD_collected["SCD"], insertionSCD_collected["SCD"]).statistic, 4)
plt.title(f"Average SCD correlation between insertion and disruption exps, corr={perm_corr}")
plt.xlabel("aveSCD insertion")
plt.ylabel("aveSCD disruption")
```
