## Correlation between insertion and disruption SCD

```python
fig, ax = plt.subplots(figsize=(8, 8))

plt.scatter(x=disruption_df[stat_of_analysis_interest], y=insertion_df[stat_of_analysis_interest], alpha=0.3)
perm_corr = round(stats.pearsonr(disruption_df[stat_of_analysis_interest], insertion_df[stat_of_analysis_interest]).statistic, 4)

lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]

# # now plot both limits against eachother
# ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)

plt.xlabel("Disruption Score")
plt.ylabel("Insertion Score")

plt.text(x=0.71, y=0.05, s=f"Pearson R = {perm_corr}", transform=plt.gca().transAxes, verticalalignment='top')

# plt.savefig("./plots/disruption_score_insertion_score.pdf", format="pdf", bbox_inches="tight")
```

## Coloring by number of sites it the boundary

```python
boundaries_file = "/project/fudenber_735/GEO/bonev_2017_GSE96107/distiller-0.3.1_mm10/results/coolers/features/bonev2017.HiC_ES.mm10.mapq_30.1000.window_200000.insulation"
```

```python
boundaries = pd.read_csv(boundaries_file, sep="\t")
```

```python
window_size = boundaries_file.split("window_")[1].split(".")[0]
boundary_key, insulation_key = (
    f"boundary_strength_{window_size}",
    f"log2_insulation_score_{window_size}",
)
```

```python
boundary_strength_thresh = 0.25
boundary_insulation_thresh = 0
```

```python
boundaries = boundaries.iloc[
        (boundaries[boundary_key].values > boundary_strength_thresh)
        * (
            boundaries[insulation_key].values
            < boundary_insulation_thresh
        )
    ]
```

```python
len(boundaries)
```

```python
half_resolution = 10000
```

```python
len(disruption_df)
```

```python
num_sites_in_boundary = []

for s in disruption_df.itertuples():
    chrom = s.chrom 
    start = s.start 
    end = s.end
    CTCF_in_boundary = disruption_df[(disruption_df["chrom"] == chrom) & (disruption_df["start"] >= start - half_resolution) & (disruption_df["end"] <= end + half_resolution)].sort_values(by="start", ascending=False)
    CTCF_in_boundary = CTCF_in_boundary[CTCF_in_boundary["SCD"] > 10]
    num_sites_in_boundary.append(len(CTCF_in_boundary))
```

```python
disruption_df["num_sites"] = num_sites_in_boundary
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

sns.scatterplot(x=disruption_df[stat_of_analysis_interest], y=insertion_df[stat_of_analysis_interest], alpha=0.5, hue=disruption_df["num_sites"])
perm_corr = round(stats.pearsonr(disruption_df[stat_of_analysis_interest], insertion_df[stat_of_analysis_interest]).statistic, 4)

lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]

# # now plot both limits against eachother
# ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)

plt.xlabel("Disruption Score")
plt.ylabel("Insertion Score")

plt.text(x=0.71, y=0.05, s=f"Pearson R = {perm_corr}", transform=plt.gca().transAxes, verticalalignment='top')
```
